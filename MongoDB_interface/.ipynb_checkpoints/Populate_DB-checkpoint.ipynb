```python
from pymongo import MongoClient
import pandas as pd
```

```python
#Getting the database
dataset = pd.read_csv('aapl_msi_sbux.csv')
```

```python
#Extracting the indexes
indexes = list(range(dataset.shape[0]))
```

```python
#Client object
client = MongoClient()
#Mongo runs on port number 27017
client = MongoClient('localhost',27017)
#stock_type
type_stock = ["COMMON_STOCK","PREFFERED_STOCKS"]
```

```python
from random import sample
```

```python
#Making a database
db = client.Apple
#Creating a collection
stock_prices = db.stock_prices
for i in range(100):
    for x in indexes:
        stock_prices.insert(
            {
                'day':x,
                'stock_price':dataset.iloc[x,0] * 72.3
                'stock_type': type_stock[0]
            }
        )
        stock_prices.insert(
            {
                'day':x,
                'stock_price':dataset.iloc[x,0] * 72.3
                'stock_type': type_stock[1]
            }
        )
#Displaying the collections
for x in stock_prices.find():
    print(x)
```

```python
#Making a database
db = client.MSI
#Creating a collection
stock_prices = db.stock_prices
for i in range(100):
    for x in indexes:
        stock_prices.insert(
            {
                'day':x,
                'stock_price':dataset.iloc[x,1] * 72.3
                'stock_type': type_stock[0]
            }
        )
        stock_prices.insert(
            {
                'day':x,
                'stock_price':dataset.iloc[x,1] * 72.3
                'stock_type': type_stock[1]
            }
        )
    
#Displaying the collections
for x in stock_prices.find():
    print(x)
```
